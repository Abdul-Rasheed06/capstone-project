# Imputing The remainig missing values

# KNN imputer method is always appreciated as it's advanced method

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=11) # You can always go odd number to better results min=5 max=any odd value
company_impute = pd.DataFrame(imputer.fit_transform(company_sub2), columns = company_sub2.columns)
```

# inspect possible correlations between independent variables

```python
# Splitting data into ind and dep for building model and prediction
predictors =  company_impute.drop('default',axis=1)
response = company_impute['default']
```

# Feature Extraction Method

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
# split the data into train and test
x_train, x_test, y_train, y_test = train_test_split(predictors, response, test_size=0.3, random_state=2)
```

# For modeling we will use Logistic Regression with recursive elimination method


```python
logitR = LogisticRegression()
selector = RFE(estimator = logitR,  n_features_to_select=21, step=1)
selector = selector.fit(x_train, y_train)
```

```python
selector.n_features_
```

```python
selector.ranking_
```

```python
x_train.columns
```

```python
df = pd.DataFrame({'Feature':scaled_predictors.columns, 'Rank':selector.ranking_})
df[df['Rank']==1]
```

# Validating The model on train and test

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
pred_train = selector.predict(x_train)
pred_test = selector.predict(x_test)
```

```python
# Evaluate the model Preformence
print(confusion_matrix(y_train, pred_train))
print()
print(confusion_matrix(y_test, pred_test))
```

```python
print(classification_report(y_train, pred_train))
print()
print(classification_report(y_test, pred_test))
```

```python
print("Training Accuracy :", accuracy_score(y_train, pred_train))
print()
print("Testing Accuracy :",accuracy_score( y_test, pred_test))
```

- But Here is a probelm we dont handle the imbalance treatment saw it is always appreciated the handle imbalance treatment

# SMOTE Method

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=33)
x_smote,y_smote = smote.fit_resample(x_train, y_train) # here we consider the x_train, y_train to solve the Data Leakage problem
```

```python
print(x_train.shape, y_train.shape, x_smote.shape, y_smote.shape)
```

```python
print("Imbalance Data :", y_train.value_counts())
print("balanced Data :", y_smote.value_counts())
```

```python
selector_smote = selector.fit(x_smote, y_smote)
```

```python
selector_smote.n_features_
```

```python
pred_train_smote = selector_smote.predict(x_smote)
pred_test_smote = selector_smote.predict(x_test)
```

```python
# Evaluate the model Preformence
print(confusion_matrix(y_smote, pred_train_smote))
print()
print(confusion_matrix(y_test, pred_test_smote))
```

```python
print(classification_report(y_smote, pred_train_smote))
print()
print(classification_report(y_test, pred_test_smote))
```

```python
print("Training Accuracy :", accuracy_score(y_smote, pred_train_smote))
print()
print("Testing Accuracy :",accuracy_score( y_test, pred_test_smote))
```

# AUC and ROC

```python
from sklearn.metrics import roc_auc_score
logistic_roc_auc = roc_auc_score(y_test, pred_test_smote)
logistic_roc_auc
```

```python
from sklearn.metrics import roc_curve
fp, tpr, thresholds = roc_curve(y_test, pred_test_smote)
plt.figure(figsize=(8,8))
plt.plot(fp, tpr, label = "ROC Curve (area=0.2f)" %logistic_roc_auc)
plt.plot([0,1], [0,1], 'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Reciever Oprating Characteristic With Area Under Curve")
plt.legend(loc="lower right")
plt.show()
```
