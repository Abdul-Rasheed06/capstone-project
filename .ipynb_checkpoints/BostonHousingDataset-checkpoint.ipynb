# preprocessing part

```python
# 1) handling missing value
# 2) encoding concept
# 3) outlier treatement
# 4) feature scaling
```

```python
# no missing data found
df.isnull().sum().sum()
```

```python
# encoding is not required
df.info()
```

```python
# 3) outlier 
# crime, zn, rm,b,
```

```python
df['crim'].describe()
```

```python
Q1 = 0.08
Q3 = 3.67
IQR = Q3 - Q1
upper_limit = Q3 + 1.5 * IQR
df['crim'] = np.where(df['crim']>upper_limit, upper_limit,df['crim'])
```

```python
df['zn'].describe()
```

```python
Q1 = 0.000000
Q3 = 12.500000
IQR = Q3 - Q1
upper_limit = Q3 + 1.5 * IQR
df['zn'] = np.where(df['zn']>upper_limit, upper_limit,df['zn'])
```

```python
df['b'].describe()
```

```python
Q1 = 375.377500
Q3 = 396.225000
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
df['b'] = np.where(df['b']<lower_limit, lower_limit,df['b'])
```

```python
# check - boxplot
def boxplots(col):
    sns.boxplot(df[col])
    plt.show()

for i in list(df.select_dtypes(exclude=['object']).columns)[0:]:
    boxplots(i)
```

```python
# split the data into ind and dep variable
x = df.drop(['medv'], axis=1) # independent variable
#x = df.iloc[:,:-1]
y = df['medv'] # dependent variable
```

```python
x.head()
```

```python
y
```

# feature scaling 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
x_scaler = pd.DataFrame(x_scaler, columns=x.columns)
```

```python
x_scaler.head()
```

```python
x_scaler.describe().T.round(2)
```

# 'VIF - variance inflation factor > 5' - multicollinearity

```python
x_scaler.shape
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variable = x_scaler
vif = pd.DataFrame()
vif['Variance Inflation Factor'] = [variance_inflation_factor(variable, i) for i in range(variable.shape[1])]
vif['Features'] = x.columns                                    
```

```python
vif
```

```python
# rad has a highest vif = 12.657724, hence dropping this variable
x_scaler = x_scaler.drop('rad', axis=1)
x_scaler.columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variable = x_scaler
vif = pd.DataFrame()
vif['Variance Inflation Factor'] = [variance_inflation_factor(variable, i) for i in range(variable.shape[1])]
vif['Features'] = x_scaler.columns  
vif
```
