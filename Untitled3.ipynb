```python
# Before do VIF we Have to handle the outlers
df['crim'].describe()
```

```python
# IQR Method
Q1 =  0.082045
Q3 = 3.677083
IQR = Q3-Q1
upper_limit = Q3+1.5*IQR
lower_limit = Q1-1.5*IQR

df['crim'] = np.where(df['crim'] > upper_limit,upper_limit,
                     np.where(df['crim'] < lower_limit,lower_limit,df['crim']))
```
